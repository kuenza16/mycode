Task 1: Postman Versions

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("dataset\\billboard_charts.csv")
data
data = data[data["song"]=="Please Mr. Postman"]
data = data.groupby(['artist'])['artist'].count()
data
explode =[0.1,0,0]
plt.pie(data, labels=data.index, explode=explode,autopct='%1.1f%%')
```

Task 2: School District Cost Breakdown

```python
import matplotlib.pyplot as plt

labels = [ "Building Construction", "Deferred Maintenance", "General Contractor Fees", "Owner Costs", "Site Construction" ]
costs = [47945975,  3753584, 7340612,  18247137, 576668]
colors = ["#355070", "#6d597a", "#b56576", "#e56b6f", "#eaac8b"]

explode = (0, 0.3,0,0,0.3)  

fig1, ax1 = plt.subplots()
ax1.pie(costs, explode=explode, labels=labels, autopct='%1.1f%%',colors=colors,
        shadow=True, startangle=90)
ax1.axis('equal')  

plt.show()
```
