Task 1: Billboard chart

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("dataset\\billboard_charts.csv")
data = data[['artist','weeks-on-board','peak-rank']]
data = data[data['peak-rank']==1]
data.sort_values(["weeks-on-board",'peak-rank'], axis = 0, ascending = True,
                 inplace = True, na_position ='last')

data=data.drop_duplicates(subset='artist', keep="last")
data = data.tail(10)
x = data['weeks-on-board']
y = data['artist']
plt.barh(y, x, color='r')
plt.title("innovative companies")
plt.xlabel("Countries")
plt.ylabel("Number of Companies")

plt.show()
```

Task 2: Snow storm report

```python
import matplotlib.pyplot as plt

resorts = ["Crested Butte", "Aspen", "Telluride", "Wolf Creek"]
snow_last_24 = [29,15,15,23]
snow_next_24 = [16,6,19,30]

plt.title("CO Snow Strom Report")
plt.ylabel("Inches")
plt.bar(resorts, snow_last_24, color='#74b9ff')

plt.bar(resorts, snow_next_24, bottom=snow_last_24, color='#7ed6df')
plt.legend(["Last 24hrs","Next 24hrs"])
plt.show()
```
