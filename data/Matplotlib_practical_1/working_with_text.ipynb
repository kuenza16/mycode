Task 1: Add text and “text-in-box”

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0.0,5.0,0.01)
y = np.cos(2*np.pi*x)


fig=plt.figure(1)
ax = fig.add_subplot(111)
ax.text(2,1,'x=2, y=1',style='italic')
ax.text(5,1,'x=5, y=1',style='italic',bbox={'facecolor':'red','alpha':0.5})


plt.plot(x,y)
plt.show()
```

Task 2: Add annotation (arrow)

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0.0,5.0,0.01)
y = np.cos(2*np.pi*x)

fig=plt.figure(1)
ax = fig.add_subplot(111)
plt.ylim(-2, 2)
plt.xlim(0,5)
ax.annotate('x=2, y=1', xy=(2, 1),  xycoords='data',
            xytext=(0.3, 0.95), textcoords='axes fraction',
            arrowprops=dict(facecolor='blue', shrink=0.05),
            horizontalalignment='right', verticalalignment='top',
            )
ax.annotate('x=1.5, y=-1', xy=(1.5, -1),  xycoords='data',
            xytext=(0.2, -0.009), textcoords='axes fraction',
            arrowprops=dict(facecolor='red', shrink=0.07),
            horizontalalignment='right', verticalalignment='bottom',
            )


plt.plot(x,y)
plt.show()
```

Task 3: Add axis labels, subtitles and legend

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 128)
y1 = np.cos(x) # y values for c1
y2 = np.sin(x) # y values for c2

fig = plt.figure(figsize=(3, 3))
ax = fig.add_subplot(111)
plt.title("Working with text demo")
ax.plot(x, y1)
ax.plot(x, y2)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.legend(['c1','c2'])
```
